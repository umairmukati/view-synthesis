```python
def single_run(index, img_path = img_paper, img_name = 'Seahorse.png'):
    
    p[0] = (index[0] - lfsize[2]//2)/(lfsize[2]//2)
    q[0] = (index[1] - lfsize[3]//2)/(lfsize[3]//2)
        
    img = cv2.imread(os.path.join(img_path, img_name))
                
    img = processLF(trans(img), lfsize, gamma_val)

    pers = img[100:292, 200:392, index[0], index[1], :].squeeze()
    corn = img[100:292, 200:392, [0, -1, 0, -1], [0, 0, -1, -1], :].squeeze()

    corn = corn.permute(2,3,0,1).reshape(12,pers.shape[0],pers.shape[1])[None,:]
    pers = pers[None,:].permute(0,3,1,2)
    
    with torch.no_grad():
        O_view = net(get_variable(corn), get_variable(torch.from_numpy(p)), get_variable(torch.from_numpy(q)))
        net_out = get_numpy(O_view)
        Y = get_numpy(pers)

    return net_out, Y
```

```python
def synthesizeView(corn, index):

    p[0] = (index[0] - lfsize[2]//2)/(lfsize[2]//2)
    q[0] = (index[1] - lfsize[3]//2)/(lfsize[3]//2)
    
    corn = corn.permute(2,3,0,1).reshape(12,corn.shape[0],corn.shape[1])[None,:]

    with torch.no_grad():
        O_view = net(get_variable(corn), get_variable(torch.from_numpy(p)), get_variable(torch.from_numpy(q)))
        
    return O_view.squeeze().permute(1,2,0)
```

```python
# Run for all examples in the folder and every perspectives
def run_all_examples(img_path):
    
    
    files = [file for file in os.listdir(img_path) if file.endswith(".png")]
    ps = np.ndarray((len(files),7,7))

    for fi in range(len(files)):

        file = files[fi]

        print("Current file {}: {}" .format(fi,file))

        img = cv2.imread(os.path.join(img_path, file))
        img = processLF(trans(img), lfsize, gamma_val)

        corn = img[100:292, 200:392, [0, -1, 0, -1], [0, 0, -1, -1], :].squeeze()
        
        
        pdb.set_trace()

        for i in range(7):
            for j in range(7):

                T = get_numpy(img[100:292, 200:392, i, j, :])
                
                p[0] = (i - lfsize[2]//2)/(lfsize[2]//2)
                q[0] = (j - lfsize[3]//2)/(lfsize[3]//2)
                
                with torch.no_grad():
                    O_view = net(X_corners[7][None,:], get_variable(torch.from_numpy(p)), get_variable(torch.from_numpy(q)))
                    
                Y = O_view.squeeze().permute(1,2,0)
                
                #Y = get_numpy(synthesizeView(x, [i, j]))

                ps[fi,i,j] = psnr_1(T, Y)

        print("Current PSNR: {}" .format(ps[fi].mean()))
    
    return ps
                
    
```
